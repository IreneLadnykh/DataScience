```python
#13 Количество игр, которая сыграла каждая команда в сезоне
game['away_team'].value_counts() + game ['home_team'].value_counts()
```

```python
#14 Какая команда выиграла наибольшее количество матчей в сезоне
game['delta_run']=game['away_team_runs'].sub(game['home_team_runs'])
game['win_team']=np.where(game['delta_run']>0, game['away_team'],game['home_team'])

win_count=game['win_team'].value_counts(ascending=False)
print (win_count)
```

```python
#15 Какая команда выиграла наибольшее количество домашних матчей
game['win_home_team']=np.where(game['home_team_win']==1, game['home_team'],0)
game['win_home_team'].value_counts(ascending=False)
```

```python
#16 Какая команда выиграла наибольшее количество гостевых матчей
game['delta_run']=game['away_team_runs'].sub(game['home_team_runs'])
game['win_away_team']=np.where(game['delta_run']>0, game['away_team'],0)

win_away_count=game['win_away_team'].value_counts(ascending=False)
print (win_away_count)
```

```python
#17 Какая команда проиграла наибольшее количество матчей в сезоне
game['delta_run']=game['away_team_runs'].sub(game['home_team_runs'])
game['loss_team']=np.where(game['delta_run']>0, game['home_team'],game['away_team'])

win_count=game['loss_team'].value_counts(ascending=False)
print (win_count)
```

```python
#18 Зависит ли выигрыш от количества посетителей матча. Я так поняла, что выигрыш домашней команды. Задание сформулировано расплывчато
game['cov']=game['attendance'].cov(game['home_team_win'])
print (game['cov'])
```

```python
#19 Правда ли что большинство проигрышей было в субботу. Я так поняла, что проигрыш домашней команды.
game['delta_run']=game['away_team_runs'].sub(game['home_team_runs'])
game['loss_team']=np.where(game['delta_run']>0, 1,0)
game['loss_day']=np.where(game['loss_team']==1, game['day_of_week'],0)
game['loss_day_number']=np.where(game['day_of_week']=='Saturday',1,0)
game['loss_day_number']=np.where(game['day_of_week']=='Sunday',1,0)
sum_loss=game['loss_day_number'].sum()
total_loss=game['loss_team'].sum()
#Вывожу соотношение проигранных матчей в субботу и воскресенье к общему количеству проигранных домашних матчей
print (sum_loss/total_loss)
```

```python
#20 Правда ли, что наибольшее количество ранов происходит в субботу
game['runs_saturday']=np.where(game['day_of_week']=='Saturday',game['total_runs'],0)
game['runs_others']=np.where(game['day_of_week']!='Saturday',game['total_runs'],0)
total_saturday=game['runs_saturday'].max()
total_others=game['runs_others'].max()
#Вывожу соотношение максимального количества ранов в субботу  к максимальному количеству ранов в остальные дни
print (total_saturday/total_others)
```
